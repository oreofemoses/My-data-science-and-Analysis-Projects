```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Set up the figure and axis
fig, ax = plt.subplots()

# Define a formatter function for currency formatting
def currency_formatter(value, _):
    return '${:,.0f}'.format(value)

# Format the y-axis as currency using the formatter function
ax.yaxis.set_major_formatter(FuncFormatter(currency_formatter))

# Plot the bar graph
bars = ax.bar(np.arange(len(df)), df['Profit'], label='Profit')

# Add data labels to the bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 500, str(df['Year'][i]),
            ha='center', va='bottom', fontsize=8)

# Set x-axis tick positions and labels
ax.set_xticks(np.arange(len(df)))
ax.set_xticklabels(df['Year'])

# Set axis labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Profit')
ax.set_title('Profit by Year')

# Add legend
ax.legend()

# Show the plot
plt.xticks(())
plt.tight_layout()
plt.savefig('bar_graph.jpg', dpi=300)
plt.show()

```
